```python
train = pd.read_csv("C:/Users/admin/Desktop/DataCastle-Solution-master/code/data/train_x_rank.csv")
train.head()
```

```python
#去掉uid这一列
train_x = train.drop(['uid'],axis=1)
train_x.head()
```

```python
test = pd.read_csv("C:/Users/admin/Desktop/DataCastle-Solution-master/code/data/test_x_rank.csv")
test.head()
```

```python
test_x = test.drop(['uid'],axis=1)
test_x.head()
```

```python
train_unlabeled =  pd.read_csv("C:/Users/admin/Desktop/DataCastle-Solution-master/code/data/train_unlabeled_rank.csv")
train_unlabeled.head()
```

```python
train_unlabeled_x =  train_unlabeled.drop(['uid'],axis=1)
```

```python
#阈值离散切分
train_x[train_x<1500] = 1
train_x[(train_x>=1500)&(train_x<3000)] = 2
train_x[(train_x>=3000)&(train_x<4500)] = 3
train_x[(train_x>=4500)&(train_x<6000)] = 4
train_x[(train_x>=6000)&(train_x<7500)] = 5
train_x[(train_x>=7500)&(train_x<9000)] = 6
train_x[(train_x>=9000)&(train_x<10500)] = 7
train_x[(train_x>=10500)&(train_x<12000)] = 8
train_x[(train_x>=12000)&(train_x<13500)] = 9
train_x[train_x>=13500] = 10

train_x.head()
```

```python
#离散特征的命名：在原始特征前加'd',如'x1'的离散特征为'dx1'
rename_dict = {s:'d'+s[1:] for s in train_x.columns.tolist()}
rename_dict
```

```python
train_x.head()
```

```python
#查看列名
train_x.columns.tolist()
```

```python
#替换列名
train_x = train_x.rename(columns=rename_dict)
train_x.head()
```

```python
train_x.columns.tolist()
```

```python
train_x['uid'] = train.uid
```

```python
train_x.head()
```

```python
#保存到train_x_discretization.csv
train_x.to_csv('C:/Users/admin/Desktop/DataCastle-Solution-master/code/data/train_x_discretization.csv',index=None)
```
